## 43.0 Solid state physics: Lattice systems in 3D

```python
a = 1   #lattice constant
basis = np.array([[0,0,0],[a/4,a/4,a/4]])  #atom in the base
neighbours = [-2,-1, 0, 1, 2]   #maxium displacement, where 2 means 2*[v1,v2,v3]
ai = primitive_vectors("fcc", a, c=2)  #primitive vectors of well-known systems (fcc, hcp, sc, bcc)
#ai = np.array([1,-1,0],[1,1,0],[1,2,3]) #custom primitive vectors

bi = get_recip(ai)  #receprical space lattice

points_r = []  
for base in basis:
    points_r.append(fill_space(base ,ai, neighbours))  #filling the space with nodes

points_k = fill_space(basis[0,:],bi, neighbours)  #filling receprical space 
plot_spaces(rlimit = a, klimit = 4*pi/a, points_r = points_r, points_k = points_k, ai=ai, bi=bi) #plot
```

```python
def primitive_vectors(system, a, c=2):    
    if system=="fcc":
        return a*np.array([[1,1,0],[0,1,1],[1,0,1]])/2
    if system=="bcc":
        return a*np.array([[1,1,-1],[-1,1,1],[1,-1,1]])/2
    if system=="hcp":
        return np.array([[a,0,0],[a/2,np.sqrt(3)*a/2,0],[0,0,c]])
    if system=="sc":
        return np.array([[1,0,0],[0,1,0],[0,0,1]])*a
```

```python
def get_recip(ai):
    b = np.zeros((3,3))
    V = np.dot(ai[0],np.cross(ai[1], ai[2]))
    b[0,:] = np.cross(ai[1], ai[2]) 
    b[1,:] = np.cross(ai[2], ai[0]) 
    b[2,:] = np.cross(ai[0], ai[1]) 
    b /= V/2/pi
    return b
```

```python
def fill_space(basis, vec, zones):
    vectors = np.array(list(itertools.product(zones,zones,zones)))
    points = []
    for v in vectors:
        point = basis + v[0]*vec[0,:] + v[1]*vec[1,:] + v[2]*vec[2,:]
        points.append(point)
    return np.array(points)
```

```python
def plot_spaces(rlimit, klimit, points_r, points_k, ai, bi):
    fig = plt.figure(facecolor='w', figsize=(8,16))
    ax = fig.add_subplot(2, 1, 1, projection='3d')
    bx = fig.add_subplot(2, 1, 2, projection='3d')
    colors = ['b','g']
    ax.set_title("Real Space")
    for j,base in enumerate(basis):
        ax.scatter(points_r[j][:,0],points_r[j][:,1], points_r[j][:,2], s=300, alpha = 1, c=colors[j])
    
    for n,aa in enumerate(ai):
        ax.plot([0,ai[n,0]],[0,ai[n,1]],[0,ai[n,2]],linewidth=3)
    
    ax.set_xlim(0,rlimit)
    ax.set_ylim(0,rlimit)
    ax.set_zlim(0,rlimit)
    
    
    bx.set_title("Reciprocal Space")
        
    bx.scatter(points_k[:,0],points_k[:,1], points_k[:,2], s=200, alpha = 1, c='r')
    for n,a in enumerate(ai):
        bx.plot([0,bi[n,0]],[0,bi[n,1]],[0,bi[n,2]],linewidth=3)
    
    bx.set_xlim(0,klimit)
    bx.set_ylim(0,klimit)
    bx.set_zlim(0,klimit)
    
    plt.show()
```
